## 1.1 NUMPY 

```python
import numpy as np 
# Tạo một mảng 1 chiều
mang_1D = np.array([1, 2, 3, 4, 5, 6, 7, 8])
print(mang_1D)

# Tạo một mảng 2 chiều
mang_2D = np.array([[1, 3, 5, 7], [2, 4, 6 ,8]])
print(mang_2D)

# Tạo một ngẫu nhiên
mang_nD = np.random.rand(4, 4)
print(mang_nD)
```

### Các thao tác cơ bản trong numpy 

```python
import numpy as np 
# Tạo một mảng 1 chiều
mang_1D = np.array([1, 2, 3, 4, 5, 6, 7, 8])
print(mang_1D)
#Cộng và nhân mảng với một số
print(mang_1D + 5)
print(mang_1D * 5) 

#Tính tổng trung bình, lớn nhất, nhỏ nhấtnhất
print(mang_1D.sum())
print(mang_1D.mean()) 
print(mang_1D.max())
print(mang_1D.min())   

#Lấy phần tử
print(mang_1D[2]) 
```

## Pandas

### 2.1 Tạo DataFrame

```python
import pandas as pd 

#Tạo dataframe và dictionary
data = {
    "ten": ['Ai', 'Yeu', 'Anh'],
    "Tuoi": [18, 15, 24],
    "Diem":[3, 7, 10]
}

df = pd.DataFrame(data)
print(df)
```

### 2.2 Thao tác cơ bản

```python
print(df["Tuoi"])

#Lấy hành thứ 2
print(df.iloc[1])

#Lấy ra người có tuổi lớn hơn 1818
print(df[df["Tuoi"]>18])

#Tính điểm trung bình cảu người đó
diem_trung_binh = df["Diem"].mean()
print("Điểm trung bình:", diem_trung_binh)
```

## Matplotlib

### 3.1 Biểu đồ đường

```python
import matplotlib.pyplot as plt 

x = [5, 8, 7, 4, 32, 99]
y = [20, 87, 64, 55, 30, 100]

plt.plot(x,y, label="Biểu đồ đường", color="pink", marker="o")
plt.title("Biểu đồ đường")
plt.xlabel("Trực X")
plt.ylabel("Trực Y")
plt.legend()
plt.show()
```

### Biểu đồ cột

```python
tên = ["A", "B", "C", "D", "E"]
điểm = [2, 2.5, 1, 1.5, 0.5]

plt.bar(tên, điểm, color="orange")
plt.title("Biểu đồ cột")
plt.xlabel("têntên")
plt.ylabel("điểmđiểm")
plt.legend()
plt.show()
```

### Biểu đồ phân tán

```python
x = np.random.rand(100)
y = np.random.rand(100)

sizes = np.random.rand(100)*100

plt.scatter(x, y, s=sizes, alpha=0.5, c="blue")
plt.title("Biểu đồ phân tán")
plt.xlabel("Trực X")
plt.ylabel("Trực Y")
plt.legend()
plt.show()
```
